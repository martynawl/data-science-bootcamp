Tworzenie wielu wykresów

```python
plt.subplot(121) # co kolejny oznacza: 1-wierszy, ile chcemy uzyskać w naszej siatce wykresów, 
#2-określa lioczbę kolumn, ile chcemy uzyskać w naszej siatce wykresów, 
#1- odnosi się do konkretnego wykresu - do pierwszego wykresu
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 2, 4], 'r--')
```

Matplotlib: wykresy słupkowe i punktowe

Wykresy słupkowe

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2]) #wykres wertykalny
```

```python
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2]) #wykres horyzontalny
```

Wykresy punktowe

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300) * 50

plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5) #alpha oznacza przezroczystość
```

Matplotlib: wyświetlanie obrazów, subploty

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('office.jpg')
img
```

```python
img.shape
```

```python
plt.imshow(img)
_ = plt.axis('Off')
```

```python
fig = plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
```

```python
#Automatyzujemy, to co powyżej (żeby nie pisać tyle kodu)

fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(140 + idx + 1)
    plt.imshow(mpimg.imread(img))     
```

Matplotlib: style

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)

plt.plot(x, x_sin)
```

```python
plt.style.use('ggplot')
plt.plot(x, x_sin)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-ticks')
plt.plot(x, x_sin)
```

```python
plt.style.use('dark_background')
plt.plot(x, x_sin)
```

```python
plt.style.use('seaborn')
plt.plot(x, x_sin)
```
