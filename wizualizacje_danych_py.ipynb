```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
plt.plot(x, x_cos)
```

Tworzenie wielu wykresw

```python
plt.subplot(121) # co kolejny oznacza: 1-wierszy, ile chcemy uzyska w naszej siatce wykresw, 
#2-okrela lioczb kolumn, ile chcemy uzyska w naszej siatce wykresw, 
#1- odnosi si do konkretnego wykresu - do pierwszego wykresu
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 2, 4], 'r--')
```

```python
flg = plt.figure(figsize=(12, 8))

plt.subplot(131) 
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 2, 4], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 2, 4], 'ro')

```

Matplotlib: wykresy supkowe i punktowe

Wykresy supkowe

```python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
```

```python
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2]) #wykres wertykalny
```

```python
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2]) #wykres horyzontalny
```

Wykresy punktowe

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
```

```python
x3 = np.random.randn(300) * 50

plt.scatter(x=x1, y=x2, s=x3)
```

```python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5) #alpha oznacza przezroczysto
```
